```python
import pandas as pd
import os
from datetime import datetime
```

```python
data_path = "Trending-YouTube-Scraper-master/"
```

```python
file_list = os.listdir(data_path)
csv_list = []
for file in file_list:
    if file.endswith(".csv"):
#         file_name = "\"" + file + "\""
        csv_list.append(file)
```

```python
csv_list
```

```python
ytb_dfs = []

for file in csv_list:
    df = pd.read_csv(data_path + file)
    ytb_dfs.append(df)
```

```python
yt_df = pd.concat(ytb_dfs)
yt_df = yt_df.reset_index(drop=True)
yt_df
```

```python
yt_df.info()
```

```python
yt_df['categoryId'].unique()
```

```python
category_dict = {1: "Film & Animation", 2: "Autos & Vehicles", 10: "Music", 15: "Pets & Animals", 17: "Sports", 19: "Travel & Events",
                 20: "Gaming", 22: "People & Blogs", 23: "Comedy", 24: "Entertainment", 25: "News & Politics", 26: "Howto & Style",
                 27: "Education", 28: "Science & Technology", 29: "Nonprofits & Activism"}
```

```python
yt_df['category'] = [category_dict[k] for k in yt_df['categoryId']]
yt_df
```

```python
yt_df['trending_date'] = pd.to_datetime(yt_df['trending_date'], format = "%y.%d.%m")
```

```python
yt_df["date_added"] = yt_df['time_added'].apply(lambda x: x[:10])
# yt_df["date_added"] = pd.to_datetime(yt_df["date_added"], format = "%m/%d/%Y")
yt_df["timeadded"] = yt_df['time_added'].apply(lambda x: x[11:])
yt_df.loc[yt_df['timeadded'] == "22:30pm", "timeadded"] = "10:30pm"
```

```python
yt_df['timeadded'] = yt_df['timeadded'].apply(lambda x: datetime.strptime(x, '%I:%M%p').strftime("%H:%M"))
```

```python
yt_df.info()
```

```python
yt_df['added'] = pd.to_datetime(yt_df["date_added"] + " " + yt_df["timeadded"])
```
