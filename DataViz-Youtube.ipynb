```python
def famous_category(df):
    df = df.groupby("categoryId", as_index=False).agg({"title":"count"}).sort_values("title", ascending=False)
    df.columns = ['category', 'frequency']
    return df

def add_country(df, name):
    df['country'] = name
    return df

def add_timeadded(df):
    df['time_added'] = "04/29/2020 11:30am"
    return df
```
